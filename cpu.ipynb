```python
def fib(n):
    if n in {0,1}:
        return 1
    else:
        antes=1
        anterior=1
        for i in range(n-1):
            ahora=antes+anterior
            anterior=antes
            antes=ahora
        return ahora
```

```python
def fib2(n):
    if n in {0,1}:
        return 1
    else:
        return (fib2(n-1) + fib2(n-2))
```
